```python
import numpy as np
import ast
from collections import Counter, defaultdict
```

```python
def read_game(gamefile):
    with open(gamefile, 'r') as f:
        reader = f.read()
    l = reader.split('\n')
 
    games = [l[i:i + 2] for i in range(0, len(l), 2)]
    while not games[-1]:
        games = games[:-1]
    
    if not games[-1]:
        games = games[:-1]
    
    
    games =  [i for i in games if len(i) == 2]
    full_games = []
    for x,y in games:
        try:
            g = ast.literal_eval(y)
            full_games.append((x,g))
        except:
            pass
        
    return full_games

def win_percs(games):
    winners = [i for i,_ in games]
    w = Counter(winners)
    total = len(list(w.elements()))
    total = float(total)
    print "White win percent: ", w['W'] / total
    print "Black win percent: ", w['B'] / total
    print "Tie percent: ", w['='] / total
    print "Total games:", total
```

```python
games = read_game('client-python/pure_alpha_beta_results.txt')
win_percs(games)
```

```python
games = read_game('client-python/real_3_depth_random.txt')
win_percs(games)
```

```python
# first 2 moves for white
d = defaultdict(int)
for game in games:
    first = tuple(game[1][0][:-1])
    second = tuple(game[1][2][:-1])
    moves = (first,second)
    result = game[0]
   
    if result == 'W':
        result = 1
    elif result == 'B':
        result = -1
    else:
        result = 0
    
    d[moves] += result
```
