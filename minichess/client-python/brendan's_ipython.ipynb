```python
ab_dict = {}
for x in ab:
    fm =  x[1][:31]
    if fm in ab_dict:
        ab_dict[fm][0] += x[0]
        ab_dict[fm][1] += 1
    else:
        ab_dict[fm] = [x[0],1]
```
