```python
rl_raw = []
with open("rl_agent.txt") as f:
    for line in  f.readlines():
          rl_raw.append([line])
```

```python
rl_arr = []
for x in xrange(len(rl_raw)/9):
    mini_arr = []
    for y in xrange(9):
        mini_arr += rl_raw[9*x+y]
    rl_arr += [mini_arr]
```

```python
Counter([tuple([x[0],x[6]]) for x in rl_arr][::-1])
```

```python
rl_agent = Counter()
alpha = 0.25
for x in rl_arr:
    for y in xrange(3):
        state = None
        action = None
        if y == 0:
            state = ()
            action = x[0].strip('\n')
        elif y == 1:
            state = tuple([x[0].strip('\n')])
            action = x[2].strip('\n')
        elif y == 2:
            state = tuple([x[0].strip('\n')] + [x[2].strip('\n')])
            action = x[4].strip('\n')
        if x[6] == 'W\n' and y % 2 == 0:
            reward = 1
        elif x[6] == 'W\n' and y == 1:
            reward = -1
        elif x[6] == 'B\n' and y == 1:
            reward = 1
        elif x[6] == 'B\n' and y % 2 == 0:
            reward = -1
        elif x[6] == '=\n':
            reward = 0
        if (state,action) in rl_agent:
            rl_agent[(state, action)] = rl_agent[(state, action)] +  alpha*(reward - rl_agent[(state, action)])
        else:
            rl_agent[(state, action)] = alpha*reward
```
